```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Creating a sample dataset (simulating a diabetes dataset)
data = {
    'Pregnancies': [2, 4, 1, 3, 5, 0, 6, 1, 8, 2],
    'BloodPressure': [72, 80, 65, 75, 85, 60, 90, 70, 88, 76],
    'SkinThickness': [35, 40, 30, 33, 36, 28, 45, 31, 50, 38],
    'Insulin': [120, 130, 100, 110, 150, 90, 160, 105, 200, 140],
    'BMI': [32.0, 34.2, 28.5, 30.1, 35.0, 25.5, 38.5, 27.8, 41.0, 36.2],
    'DiabetesPedigreeFunction': [0.372, 0.245, 0.652, 0.125, 0.500, 0.187, 0.785, 0.233, 0.900, 0.412],
    'Age': [25, 30, 22, 28, 35, 21, 40, 24, 50, 32],
    'Glucose': [85, 98, 75, 90, 105, 70, 120, 80, 140, 100]  # Target variable
}

# Convert data dictionary to a Pandas DataFrame
df = pd.DataFrame(data)

# Define feature matrix (X) by dropping the 'Glucose' column
X = df.drop(columns=['Glucose'])

# Define target variable (y) as the 'Glucose' column
y = df['Glucose']

# Split the dataset into training (70%) and testing (30%) sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize the Linear Regression model
model = LinearRegression()

# Train the model
model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = model.predict(X_test)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
accuracy = r2 * 100

# Print evaluation metrics
print(f"Mean Squared Error: {mse:.2f}")
print(f"R-Squared: {r2:.2f}")
print(f"Model Accuracy: {accuracy:.2f}%")

# Create a comparison DataFrame
comparison_df = pd.DataFrame({'Actual': y_test.values, 'Predicted': y_pred})

# Visualization - Bar Chart
plt.figure(figsize=(10, 5))
plt.bar(range(len(y_test)), y_test, color='blue', alpha=0.6, label='Actual')
plt.bar(range(len(y_pred)), y_pred, color='red', alpha=0.6, label='Predicted')
plt.xlabel('Samples')
plt.ylabel('Glucose Level')
plt.title('Actual vs Predicted Glucose Levels')
plt.legend()
plt.show()

# Visualization - Scatter Plot
plt.figure(figsize=(8, 5))
plt.scatter(y_test, y_pred, color='blue', label='Predictions')
plt.plot(y_test, y_test, 'r--', label='Perfect Fit')  # Red dashed line for perfect predictions
plt.xlabel('Actual Glucose Levels')
plt.ylabel('Predicted Glucose Levels')
plt.title('Actual vs Predicted Glucose Scatter Plot')
plt.legend()
plt.show()
```
